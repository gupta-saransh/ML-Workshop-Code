```python
import pandas as pd
import numpy as np
import bs4
import urllib.request as url
```

```python
page=url.urlopen('http://stats.espncricinfo.com/india/engine/records/batting/most_runs_career.html?class=1;id=6;type=team')
```

```python
print(page)
```

```python
soup = bs4.BeautifulSoup(page,'lxml')
```

```python
data = soup.find_all('table', class_='engineTable')
```

```python
# Check what data is coming
for t in data:
    print(t.text)
```

```python
dataset = []
for t in data:
    dataset.append(t.text)
```

```python
# print(dataset)
final_data=[]
final_data=dataset[0].split('\n\n')
```

```python
del final_data[0]
ult_data=[]
for row in final_data:
    ult_data.append(row.split('\n'))
df = pd.DataFrame(ult_data)
```
